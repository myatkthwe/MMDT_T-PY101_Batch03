Ex 3: Write a function called number_of_factors that takes an integer and returns how many
factors the number has.

```python
def number_of_factors(n):
    count = 0
    for i in range(1, int(n**0.5) + 1):
        if n % i == 0:
            if i * i == n:
                count += 1
            else:
                count += 2
    return count

print(f"Number of factors for 12: {number_of_factors(12)}")
print(f"Number of factors for 49: {number_of_factors(49)}")
```

Ex4: Write a function called first_diff that is given two strings and returns the first location in
which the strings differ. If the strings are identical, it should return -1.Ignore the case in your function. It means h and H are the same.

```python
def first_diff(s1, s2):

    s1_lower = s1.lower()
    s2_lower = s2.lower()

    # Get the minimum length to iterate over
    min_len = min(len(s1_lower), len(s2_lower))

    # Iterate through the strings up to the length of the shorter string
    for i in range(min_len):
        if s1_lower[i] != s2_lower[i]:
            return i  # Return the first index where they differ

    # If one string is a prefix of the other (e.g., 'test' vs 'testing')and all characters up to min_len are the same, the difference is at min_len
    if len(s1_lower) != len(s2_lower):
        return min_len

    # If both strings are identical in content and length
    return -1

print(f"first_diff('short', 'longerstring'): {first_diff('short', 'longerstring')}")
print(f"first_diff('kay', 'KAY'): {first_diff('kay', 'KAY')}")
print(f"first_diff('applepie', 'Apple'): {first_diff('applepie', 'Apple')}")
```

Ex6: Write a function that solves the system of equations ax + by = e and cx +dy = f. Your function should takes 6 parameters, (a, b, c, d, e,f) and return the values of x and y.

```python
def solve_system(a, b, c, d, e, f):
    """
    Solves the system of equations:
        a*x + b*y = e
        c*x + d*y = f
    Returns (x, y) as a tuple.
    """
    det = a * d - b * c
    if det == 0:
        raise ValueError("The system has no unique solution (determinant is zero).")

    x = (e * d - b * f) / det
    y = (a * f - e * c) / det

    return x, y

result = solve_system(2, 3, 1, -1, 1, -3)
print("x =", result[0], "y =", result[1])
```

Ex7:  Write a function called change_case that given a string, returns a string with each upper  case letter replaced by a lower case letter and vice-versa

```python
def change_case(input_string):
    result = []
    for char in input_string:
        if 'a' <= char <= 'z':
            result.append(char.upper())
        elif 'A' <= char <= 'Z':
            result.append(char.lower())
        else:
            result.append(char)
    return "".join(result)


print(f"change_case('hello kay'): {change_case('hello kay')}")
print(f"change_case('PYTHON101'): {change_case('PYTHON101')}")
```

Ex8:  Write a function called primes that is given a number n and returns a list of the first n  primes. Let the default value of n be 100.

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def primes(n=100):
    prime_list = []
    num = 2
    while len(prime_list) < n:
        if is_prime(num):
            prime_list.append(num)
        num += 1
    return prime_list

print(f"First 5 primes: {primes(5)}")
print(f"First 10 primes: {primes(10)}")
print(f"First 100 primes (default): {primes()}")
```

Ex9: A Tic-tac-toe board can be represented be a 3×3 two-dimensional list, where zeroes stand for  empty cells, ones stand for X’s and twos stand for O’s. Write a function that is given such a list and randomly chooses a spot in which to place  a 2 (O). The spot chosen must currently be empty and a spot must be chosen.

```python
import random

def place_o_randomly(board):
    """
    Given a Tic-tac-toe board, randomly chooses an empty spot (0) and places a 2 (O).
    Modifies the board in place.
    Raises ValueError if no empty spots are available.
    """
    empty_spots = []
    for r in range(3):
        for c in range(3):
            if board[r][c] == 0:
                empty_spots.append((r, c))

    if not empty_spots:
        raise ValueError("No empty spots left on the board.")

    chosen_spot = random.choice(empty_spots)
    board[chosen_spot[0]][chosen_spot[1]] = 2
    return board

board1 = [
        [0, 1, 0],
        [0, 0, 1],
        [0, 2, 0]
        ]

print("Original Board 1:")

for row in board1: print(row)
updated_board1 = place_o_randomly(board1)
print("\nUpdated Board 1:")

for row in updated_board1: print(row)

board2 = [
       [1, 1, 2],
       [2, 1, 1],
       [1, 2, 2]
         ]

print("\nOriginal Board 2:")

for row in board2: print(row)
try:
    place_o_randomly(board2)
except ValueError as e:
    print(f"\nError: {e}")
```

Ex10:  Write a function that is given a 9 × 9 potentially solved Sudoku and returns True if it is solved correctly and False if there is a mistake. The Sudoku is correctly solved if there are
 no repeated numbers in any row or any column or in any of the nine “blocks.”


```python
def is_valid_unit(unit):
    # Filter out zeros (empty cells) before checking for duplicates
    # Sudoku numbers are 1-9. Zeros are placeholders for empty cells.
    nums = [n for n in unit if n != 0]
    # A unit is valid if all non-zero numbers are unique
    return len(nums) == len(set(nums))

def is_sudoku_solved(board):
    # 1. Check rows
    for row in board:
        if not is_valid_unit(row):
            return False

    # 2. Check columns
    for col_idx in range(9):
        column = [board[row_idx][col_idx] for row_idx in range(9)]
        if not is_valid_unit(column):
            return False

    # 3. Check 3x3 blocks
    for block_row_start in range(0, 9, 3):
        for block_col_start in range(0, 9, 3):
            block = []
            for r in range(block_row_start, block_row_start + 3):
                for c in range(block_col_start, block_col_start + 3):
                    block.append(board[r][c])
            if not is_valid_unit(block):
                return False

    # If all checks pass, the Sudoku is solved correctly
    return True

solved_board = [
      [5, 3, 4, 6, 7, 8, 9, 1, 2],
      [6, 7, 2, 1, 9, 5, 3, 4, 8],
      [1, 9, 8, 3, 4, 2, 5, 6, 7],
      [8, 5, 9, 7, 6, 1, 4, 2, 3],
      [4, 2, 6, 8, 5, 3, 7, 9, 1],
      [7, 1, 3, 9, 2, 4, 8, 5, 6],
      [9, 6, 1, 5, 3, 7, 2, 8, 4],
      [2, 8, 7, 4, 1, 9, 6, 3, 5],
     [3, 4, 5, 2, 8, 6, 1, 7, 9]
]

print(f"Is solved_board correctly solved? {is_sudoku_solved(solved_board)}")

# An incorrectly solved Sudoku board (repeated number in first row)
incorrect_board = [
     [5, 3, 4, 6, 7, 8, 9, 1, 1],
     [6, 7, 2, 1, 9, 5, 3, 4, 8],
     [1, 9, 8, 3, 4, 2, 5, 6, 7],
     [8, 5, 9, 7, 6, 1, 4, 2, 3],
     [4, 2, 6, 8, 5, 3, 7, 9, 1],
     [7, 1, 3, 9, 2, 4, 8, 5, 6],
     [9, 6, 1, 5, 3, 7, 2, 8, 4],
     [2, 8, 7, 4, 1, 9, 6, 3, 5],
     [3, 4, 5, 2, 8, 6, 1, 7, 9]
 ]

print(f"Is incorrect_board correctly solved? {is_sudoku_solved(incorrect_board)}")

# An unsolved Sudoku board (contains zeros, but otherwise valid)
unsolved_board = [
     [5, 3, 0, 6, 7, 8, 9, 1, 2],
     [6, 7, 2, 1, 9, 5, 3, 4, 8],
     [1, 9, 8, 3, 4, 2, 5, 6, 7],
     [8, 5, 9, 7, 6, 1, 4, 2, 3],
     [4, 2, 6, 8, 5, 3, 7, 9, 1],
     [7, 1, 3, 9, 2, 4, 8, 5, 6],
     [9, 6, 1, 5, 3, 7, 2, 8, 4],
     [2, 8, 7, 4, 1, 9, 6, 3, 5],
     [3, 4, 5, 2, 8, 6, 1, 7, 9]
 ]

print(f"Is unsolved_board correctly solved? {is_sudoku_solved(unsolved_board)}")
```
