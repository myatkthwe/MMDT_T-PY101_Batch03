# Ex 8: Computing.
Write a program that asks the user to enter two numbers, x and y, and computes |x-y|/(x+y).

```python
x = input ("Enter the number of x: ")
y = input ("Enter the number of y: ")

try:
  x = int(x)
  y = int(y)

  denominator = x + y
  if denominator == 0:
        print("Error: Cannot divide by zero. The sum of x and y is zero.")
  else:
        result = round(abs(x - y) / denominator,2)
        print(f"The result of |{x} - {y}| / ({x} + {y}) is: {result}")
except ValueError:
    print("Error: Invalid input. Please enter numeric values.")
```

# Ex 9: Computing

Write a program that asks the user for a number of seconds and prints out how many minutes and seconds that is. For instance, 200 seconds is 3 minutes and 20 seconds.

```python
sec = input("Enter the number of seconds: ")

try:
    total_seconds = int(sec)

    minutes = total_seconds // 60
    remaining_seconds = total_seconds % 60

    print(f"{total_seconds} seconds is {minutes} minutes and {remaining_seconds} seconds.")
except ValueError:
    print("Error: Invalid input. Please enter an integer for seconds.")
```

# Ex 10: Using Math Module

Write a program that asks the user for a number and then prints out the sine, cosine, and tangent of that number.

```python
import math
num = float(input("Enter a number:"))
sine_value = math.sin(num)
print("The sine of {} is: {}".format(num, sine_value))

cosine_value = math.cos(num)
print("The sine of {} is: {}".format(num, cosine_value))

tangent_value = math.tan(num)
print("The sine of {} is: {}".format(num, tangent_value))
```
